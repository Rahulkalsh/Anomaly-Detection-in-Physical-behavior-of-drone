```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Assuming df is your dataframe
X = df.drop("anomaly", axis=1).values
y = df["anomaly"].values

# Determine the number of timesteps and ensure compatibility
timesteps = 1  # Using 1 to ensure compatibility with 17 features
n_features = X.shape[1]

features_per_timestep = n_features // timesteps

# Ensure the total number of samples is divisible by timesteps for reshaping
n_samples = len(X) // timesteps * timesteps
X = X[:n_samples]
y = y[:n_samples]

# Reshape the data to include the timesteps dimension
X = X.reshape((n_samples // timesteps, timesteps, features_per_timestep))

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y[:len(X)], test_size=0.2, random_state=42)

# Flatten the time series data
def flatten_time_series(X):
    n_samples, timesteps, features = X.shape
    return X.reshape((n_samples, timesteps * features))

X_train_flat = flatten_time_series(X_train)
X_val_flat = flatten_time_series(X_val)

# Standardize the data
scaler = StandardScaler()
X_train_flat = scaler.fit_transform(X_train_flat)
X_val_flat = scaler.transform(X_val_flat)

# Train SVM model
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(X_train_flat, y_train)

# Evaluate SVM model
y_val_pred_svm = svm_model.predict(X_val_flat)
svm_accuracy = accuracy_score(y_val, y_val_pred_svm)
print(f"SVM Validation Accuracy: {svm_accuracy}")

# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_flat, y_train)

# Evaluate Random Forest model
y_val_pred_rf = rf_model.predict(X_val_flat)
rf_accuracy = accuracy_score(y_val, y_val_pred_rf)
print(f"Random Forest Validation Accuracy: {rf_accuracy}")
```

Bidirectional + encoder + attention

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Bidirectional, LSTM, Dropout, Dense, Concatenate, Attention, GlobalAveragePooling1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import train_test_split

# Define the encoder
encoder_inputs = Input(shape=(100, 16))
encoder_lstm = Bidirectional(LSTM(12, return_sequences=True))
encoder_outputs = encoder_lstm(encoder_inputs)
encoder_outputs = Dropout(0.2)(encoder_outputs)

# Define the decoder
decoder_inputs = Input(shape=(100, 16))  # Use the same shape as encoder inputs
decoder_lstm = Bidirectional(LSTM(12, return_sequences=True))
decoder_lstm_output = decoder_lstm(decoder_inputs)
decoder_lstm_output = Dropout(0.2)(decoder_lstm_output)

# Define the attention mechanism
attention = Attention()([decoder_lstm_output, encoder_outputs])
concat_attention = Concatenate()([decoder_lstm_output, attention])

# Add Global Average Pooling to reduce sequence to a single vector
pooled_output = GlobalAveragePooling1D()(concat_attention)

# Define the output layer
decoder_dense = Dense(1, activation='sigmoid')
decoder_outputs = decoder_dense(pooled_output)

# Create the model
BiLSTM_Attention = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
BiLSTM_Attention.compile(optimizer=Adam(learning_rate=0.001), 
                         loss='binary_crossentropy', metrics=['accuracy'])

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Assume X_train, y_train, X_val, y_val are defined and loaded properly
# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,
                                                  test_size=0.2, random_state=42)

# Prepare decoder input data (shifted target sequences)
decoder_input_data_train = np.zeros_like(X_train)
decoder_input_data_train[:, 1:, :] = X_train[:, :-1, :]  # Shift data to the right
decoder_input_data_val = np.zeros_like(X_val)
decoder_input_data_val[:, 1:, :] = X_val[:, :-1, :]  # Shift data to the right

# Measure training time
start_training_time = time.time()

# Train the model with early stopping
history = BiLSTM_Attention.fit([X_train, decoder_input_data_train], 
                               y_train, epochs=30, batch_size=32, 
                               validation_data=([X_val, decoder_input_data_val], y_val), 
                               callbacks=[early_stopping])

end_training_time = time.time()
training_time = end_training_time - start_training_time
print(f"Total training time: {training_time:.2f} seconds")

# Measure testing time
start_testing_time = time.time()

# Evaluate the model
loss, accuracy = BiLSTM_Attention.evaluate([X_val, decoder_input_data_val], y_val)

end_testing_time = time.time()
testing_time = end_testing_time - start_testing_time
print(f"Total testing time: {testing_time:.2f} seconds")

print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")

# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Bidirectional, LSTM, Dropout, Dense, Concatenate, Attention, GlobalAveragePooling1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import time

# Define the encoder
encoder_inputs = Input(shape=(64, 16))
encoder_lstm_1 = Bidirectional(LSTM(30, return_sequences=True))(encoder_inputs)
encoder_lstm_1 = Dropout(0.2)(encoder_lstm_1)
encoder_lstm_2 = Bidirectional(LSTM(15, return_sequences=True))(encoder_lstm_1)
encoder_outputs = Dropout(0.2)(encoder_lstm_2)

# Define the decoder
decoder_inputs = Input(shape=(64, 16))  # Use the same shape as encoder inputs
decoder_lstm_1 = Bidirectional(LSTM(30, return_sequences=True))(decoder_inputs)
decoder_lstm_1 = Dropout(0.2)(decoder_lstm_1)
decoder_lstm_2 = Bidirectional(LSTM(15, return_sequences=True))(decoder_lstm_1)
decoder_lstm_output = Dropout(0.2)(decoder_lstm_2)

# Define the attention mechanism
attention = Attention()([decoder_lstm_output, encoder_outputs])
concat_attention = Concatenate()([decoder_lstm_output, attention])

# Add Global Average Pooling to reduce sequence to a single vector
pooled_output = GlobalAveragePooling1D()(concat_attention)

# Define the output layer
decoder_dense = Dense(1, activation='sigmoid')
decoder_outputs = decoder_dense(pooled_output)

# Create the model
BiLSTM_Attention = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
BiLSTM_Attention.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Define learning rate scheduler
def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        # Ensure the returned value is a float
        return float(lr * tf.math.exp(-0.1))

learning_rate_scheduler = LearningRateScheduler(scheduler, verbose=1)

# Assume X_train, y_train, X_val, y_val are defined and loaded properly
# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Flatten y_train and y_val if they are 2D arrays
y_train = y_train.flatten()
y_val = y_val.flatten()

# Prepare decoder input data (shifted target sequences)
decoder_input_data_train = np.zeros_like(X_train)
decoder_input_data_train[:, 1:, :] = X_train[:, :-1, :]  # Shift data to the right
decoder_input_data_val = np.zeros_like(X_val)
decoder_input_data_val[:, 1:, :] = X_val[:, :-1, :]  # Shift data to the right

# Compute class weights
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(enumerate(class_weights))

# Measure training time
start_training_time = time.time()

# Train the model with early stopping and learning rate scheduler
history = BiLSTM_Attention.fit(
    [X_train, decoder_input_data_train],
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=([X_val, decoder_input_data_val], y_val),
    callbacks=[early_stopping, learning_rate_scheduler],
    class_weight=class_weights
)

end_training_time = time.time()
training_time = end_training_time - start_training_time
print(f"Total training time: {training_time:.2f} seconds")

# Measure testing time
start_testing_time = time.time()

# Evaluate the model
loss, accuracy = BiLSTM_Attention.evaluate([X_val, decoder_input_data_val], y_val)

end_testing_time = time.time()
testing_time = end_testing_time - start_testing_time
print(f"Total testing time: {testing_time:.2f} seconds")

print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")

# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()

# Prepare decoder input data for the test set
decoder_input_data_test = np.zeros_like(X_test)
decoder_input_data_test[:, 1:, :] = X_test[:, :-1, :]  # Shift data to the right

# Predict and adjust threshold
threshold = 0.3  # Example threshold
y_pred_prob = BiLSTM_Attention.predict([X_test, decoder_input_data_test])
y_pred = (y_pred_prob > threshold).astype(int).flatten()

# Calculate metrics
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
accuracy = accuracy_score(y_test, y_pred)
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"Accuracy: {accuracy}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Class 0', 'Class 1'])
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title('Confusion Matrix')
plt.show()
```
